# Engine Knock Prediction with an RNN

Forecast the engine knock-sensor **pressure** signal with a recurrent network (LSTM),
then **derive knock from the forecast**.

Raw cylinder-pressure data (per crank angle, per cycle) is serialized into one long
time series. An LSTM learns to predict the next stretch of pressure from a window of
past pressure samples. Knock is **not** part of the model at all - it is computed
**manually from the predicted pressure** at the very end, using a band-pass + threshold
(MAPO-style) rule.

**Pipeline**
1. **Data loading & preprocessing** - read a fuel-blend sheet, serialize it to a 1-D pressure series (no knock label, no augmentation).
2. **Inspection** - plot a slice of the serialized signal.
3. **Windowed dataset** - build sliding-window (past pressure -> future pressure) samples.
4. **Model** - an LSTM with a direct multi-step head.
5. **Training** - train and validate on a temporal split (MSE).
6. **Forecast & manual knock** - predict future pressure, then compute knock from it.

```python
import numpy as np
import pandas as pd
from scipy.signal import butter, filtfilt
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
```

```python
# ---- Data / training config ----
DATA_PATH  = "MethodsX Data 1400 rpm.xlsx"
BLEND      = "100D"      # one of: "100D", "90D10G", "80D20G", "70D30G", "60D40G"
INPUT_LEN  = 360         # context length fed to the model (= one full engine cycle)
OUTPUT_LEN = 90          # number of future steps to forecast
STRIDE     = 4           # sliding-window stride (1 = densest, larger = fewer/faster windows)
BATCH_SIZE = 128
EPOCHS     = 10
LR         = 1e-3
VAL_FRAC   = 0.2         # fraction of the (temporal) signal held out for validation

# ---- Model config (LSTM) ----
HIDDEN_SIZE = 128
NUM_LAYERS  = 2
DROPOUT     = 0.1
```

```python
# Quick environment check - tells you whether training will run on GPU or CPU.
print("torch:", torch.__version__)             # e.g. 2.x.x+cu121 (GPU build) or +cpu
print("built for CUDA:", torch.version.cuda)    # None = CPU-only build
print("CUDA available:", torch.cuda.is_available())
```

## 1. Data loading & preprocessing

`load_blend` reads one fuel-blend sheet into a `(360 crank angles, 500 cycles)` array.
`preprocess` then serializes that array **column-major**, so cycles sit end-to-end in
time order, and returns a plain 1-D pressure series of length `360 * 500 = 180000`.

There is **no knock label** (knock is derived later, from the forecast) and **no data
augmentation** - the signal is used as-is.

```python
def load_blend(blend, path=DATA_PATH):
    """Load one blend sheet (e.g. '100D', '90D10G') into a 2D numpy array.
    Returns array of shape (360 crank angles, 500 cycles)."""
    df = pd.read_excel(path, sheet_name=blend, header=None)
    return df.iloc[3:, 3:].to_numpy(dtype=float)   # (360, 500)

data = load_blend(BLEND)
```

```python
def preprocess(data):
    """Serialize (360 crank angles, 500 cycles) into a 1-D pressure series in time order.

    Each COLUMN is one cycle's crank sweep, so stacking columns end-to-end (column-major
    / Fortran order) makes every sample adjacent in time to its neighbours; the cycle
    boundaries disappear. Returns a float array of length 180000.
    """
    return data.flatten(order='F').astype(float)   # column-major (Fortran) order

pressure = preprocess(data)
```

## 2. Inspect the signal

Plot a short slice of the serialized signal (~2 engine cycles), with dashed lines at the
cycle boundaries. Use this to sanity-check the data before training.

```python
def plot_window(pressure, start, n=720):
    """Plot the serialized pressure over `n` samples starting at `start`.
    720 samples = 2 cycles (360 deg each)."""
    end = min(start + n, len(pressure))
    idx = np.arange(start, end)

    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(idx, pressure[start:end], lw=0.8, color='steelblue', label='pressure')

    # dashed lines at cycle boundaries (every 360 samples)
    for b in range((start // 360 + 1) * 360, end, 360):
        ax.axvline(b, color='gray', ls='--', lw=0.6, alpha=0.6)

    ax.set_xlabel('serialized sample index  (1 per crank angle)')
    ax.set_ylabel('knock-sensor pressure signal')
    ax.set_title(f'samples {start}\u2013{end-1}')
    ax.legend(loc='upper right')
    plt.tight_layout()
    plt.show()
    return fig
```

```python
plot_window(pressure, start=320)     # 720 samples ~ 2 full cycles
```

## 3. Windowed forecasting dataset

`PressureWindowDataset` turns the serialized signal into sliding-window samples:
- **x** = `INPUT_LEN` past steps of normalized pressure, shape `(INPUT_LEN, 1)`
- **y** = the next `OUTPUT_LEN` steps of normalized pressure, shape `(OUTPUT_LEN, 1)`

`make_loaders` splits the signal **temporally** by `VAL_FRAC` (earliest part -> train,
the rest -> val) and reuses the *training* mean/std to normalize the validation set,
avoiding leakage.

```python
class PressureWindowDataset(Dataset):
    """Sliding-window forecasting samples from a serialized pressure series.

    x: (INPUT_LEN, 1)  -> normalized past pressure
    y: (OUTPUT_LEN, 1) -> normalized pressure of the steps immediately after x
    """
    def __init__(self, pressure, input_len=INPUT_LEN, output_len=OUTPUT_LEN,
                 stride=STRIDE, mean=None, std=None):
        p = np.asarray(pressure, dtype='float32')

        # standardize pressure. Stats can be injected so the validation split reuses
        # the TRAIN statistics (no leakage).
        self.mean = float(p.mean()) if mean is None else float(mean)
        self.std  = float(p.std() + 1e-8) if std is None else float(std)
        self.p_n  = (p - self.mean) / self.std

        self.input_len = input_len
        self.output_len = output_len
        last = len(p) - input_len - output_len + 1
        self.starts = np.arange(0, max(last, 0), stride)

    def __len__(self):
        return len(self.starts)

    def __getitem__(self, idx):
        s = self.starts[idx]
        x = self.p_n[s: s + self.input_len]                                       # (L_in,)
        y = self.p_n[s + self.input_len: s + self.input_len + self.output_len]    # (L_out,)
        return torch.from_numpy(x[:, None]), torch.from_numpy(y[:, None])         # -> (L, 1)


def make_loaders(pressure, batch_size=BATCH_SIZE, val_frac=VAL_FRAC,
                 input_len=INPUT_LEN, output_len=OUTPUT_LEN, stride=STRIDE):
    """Temporal split: earliest (1-val_frac) of the signal for train, rest for val."""
    N = len(pressure)
    split = int(N * (1 - val_frac))
    train_ds = PressureWindowDataset(pressure[:split], input_len, output_len, stride)
    val_ds   = PressureWindowDataset(pressure[split:], input_len, output_len, stride,
                                     mean=train_ds.mean, std=train_ds.std)   # reuse train stats
    train_dl = DataLoader(train_ds, batch_size=batch_size, shuffle=True, drop_last=True)
    val_dl   = DataLoader(val_ds, batch_size=batch_size, shuffle=False)
    return train_dl, val_dl, (train_ds.mean, train_ds.std)
```

## 4. RNN model

A plain **LSTM** forecaster. It reads the `INPUT_LEN` window of past pressure, takes the
top layer's final hidden state, and a single linear head projects it to all `OUTPUT_LEN`
future steps at once - direct multi-step forecasting, with no autoregressive decoding.

```python
class KnockRNN(nn.Module):
    """LSTM for direct multi-step forecasting.

    Encodes the INPUT_LEN window, takes the last layer's final hidden state, and projects
    it to the full OUTPUT_LEN horizon in one shot (no autoregressive decoding).
    """
    def __init__(self, out_len=OUTPUT_LEN, hidden_size=HIDDEN_SIZE,
                 num_layers=NUM_LAYERS, dropout=DROPOUT):
        super().__init__()
        self.out_len = out_len
        self.lstm = nn.LSTM(
            input_size=1, hidden_size=hidden_size, num_layers=num_layers,
            batch_first=True, dropout=dropout if num_layers > 1 else 0.0,
        )
        self.head = nn.Linear(hidden_size, out_len)

    def forward(self, x):                 # x: (B, L_in, 1)
        _, (h, _) = self.lstm(x)          # h: (num_layers, B, hidden)
        h_last = h[-1]                    # (B, hidden) -- top layer's final hidden state
        out = self.head(h_last)           # (B, out_len)
        return out.unsqueeze(-1)          # (B, out_len, 1)
```

## 5. Training

Train with MSE loss and Adam over the temporal train/validation split, printing the
train and validation MSE each epoch. `train()` returns the fitted model, which the next
cell captures as `model` so the evaluation/plot steps below can use it.

```python
def train():
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    pressure = preprocess(load_blend(BLEND))
    train_dl, val_dl, (mu, sd) = make_loaders(pressure)

    model = KnockRNN().to(device)
    opt = torch.optim.Adam(model.parameters(), lr=LR)
    loss_fn = nn.MSELoss()

    n_params = sum(p.numel() for p in model.parameters())
    print(f"device={device}  params={n_params:,}  "
          f"train_batches={len(train_dl)}  val_batches={len(val_dl)}")

    for epoch in range(1, EPOCHS + 1):
        model.train()
        tr = 0.0
        for x, y in train_dl:
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            loss = loss_fn(model(x), y)
            loss.backward()
            opt.step()
            tr += loss.item() * x.size(0)
        tr /= len(train_dl.dataset)

        model.eval()
        va = 0.0
        with torch.no_grad():
            for x, y in val_dl:
                x, y = x.to(device), y.to(device)
                va += loss_fn(model(x), y).item() * x.size(0)
        va /= len(val_dl.dataset)
        print(f"epoch {epoch:2d}  train_mse={tr:.4f}  val_mse={va:.4f}")

    return model
```

```python
model = train()
```

## 6. Forecast & manual knock calculation

The model forecasts **pressure only**. Knock is **derived from the forecast**, manually:
the predicted pressure is band-pass filtered to isolate the high-frequency ringing and
thresholded with a MAPO-style rule (amplitude `> k x` the baseline-noise floor) inside
the post-TDC knock window.

- `evaluate` runs the model over the validation split and reports **pressure** metrics
  (MSE, MAE, R2). No knock label is involved.
- `plot_forecast` forecasts one window, then **manually computes knock** from the
  predicted pressure (and, for reference, from the true future pressure using the same
  rule) and overlays both.

```python
def _bandpass(rpm=1400, knock_band=(1200, 4000)):
    """4th-order Butterworth band-pass that isolates the knock 'ringing'.
    fs = 1 sample/deg = rpm/60*360 Hz, so Nyquist = fs/2."""
    nyq = (rpm / 60 * 360) / 2
    hi = min(knock_band[1], 0.99 * nyq)               # keep below Nyquist
    return butter(4, [knock_band[0] / nyq, hi / nyq], btype='band')


def baseline_sigma(pressure, rpm=1400, knock_band=(1200, 4000),
                   baseline_window=(-160., -40.)):
    """Noise floor: std of the band-passed signal in the knock-free baseline window
    (the same basis the threshold uses)."""
    b, a = _bandpass(rpm, knock_band)
    p = np.asarray(pressure, dtype=float)
    ca = (np.arange(len(p)) % 360) - 180              # crank angle of each sample
    base = (ca >= baseline_window[0]) & (ca <= baseline_window[1])
    return filtfilt(b, a, p)[base].std()


def knock_from_pressure(seg, start_index, sigma, knock_window=(0., 50.), k=4.0):
    """MANUAL knock detection on a pressure segment (MAPO-style).

    seg          : 1-D pressure samples
    start_index  : serialized index of seg[0]  (sets the crank-angle alignment)
    sigma        : baseline-noise std of the band-passed signal (threshold basis)
    Returns a boolean array the same length as seg: knock if the band-passed amplitude
    exceeds k*sigma AND the sample lies in the post-TDC knock window.
    """
    b, a = _bandpass()
    osc = filtfilt(b, a, seg)                          # high-frequency ringing
    idx = start_index + np.arange(len(seg))
    ca  = (idx % 360) - 180                            # crank angle of each sample
    win = (ca >= knock_window[0]) & (ca <= knock_window[1])
    return (np.abs(osc) > k * sigma) & win


@torch.no_grad()
def evaluate(model, pressure, mu, sd, val_frac=VAL_FRAC, batch_size=BATCH_SIZE):
    """Forecast the validation split, then report pressure regression metrics."""
    device = next(model.parameters()).device
    model.eval()

    split = int(len(pressure) * (1 - val_frac))
    ds = PressureWindowDataset(pressure[split:], mean=mu, std=sd)   # reuse train stats
    dl = DataLoader(ds, batch_size=batch_size, shuffle=False)

    preds, trues = [], []
    for x, y in dl:
        preds.append(model(x.to(device)).cpu().numpy()[..., 0])    # (B, OUTPUT_LEN)
        trues.append(y.numpy()[..., 0])
    preds = np.concatenate(preds)
    trues = np.concatenate(trues)
    pred_p = preds * sd + mu                                       # de-normalize
    true_p = trues * sd + mu

    mse = float(np.mean((pred_p - true_p) ** 2))
    mae = float(np.mean(np.abs(pred_p - true_p)))
    ss_res = np.sum((true_p - pred_p) ** 2)
    ss_tot = np.sum((true_p - true_p.mean()) ** 2)
    r2 = float(1 - ss_res / ss_tot) if ss_tot > 0 else float('nan')

    print(f"Pressure  MSE={mse:.4f}  MAE={mae:.4f}  R2={r2:.4f}")
    return dict(mse=mse, mae=mae, r2=r2)
```

```python
pressure = preprocess(load_blend(BLEND))     # same deterministic preprocessing as training
_, _, (mu, sd) = make_loaders(pressure)      # recover the train-set normalization stats
metrics = evaluate(model, pressure, mu, sd)
```

```python
@torch.no_grad()
def plot_forecast(model, pressure, mu, sd, start, sigma=None):
    """Forecast the OUTPUT_LEN steps after `start + INPUT_LEN`, then MANUALLY compute
    knock from the forecast (and from the true future, for reference) and overlay both."""
    device = next(model.parameters()).device
    model.eval()

    p = np.asarray(pressure, dtype='float32')
    p_n = (p - mu) / sd

    # 1) model forecast (pressure only)
    x = p_n[start:start + INPUT_LEN][:, None]                     # (INPUT_LEN, 1)
    pred_n = model(torch.from_numpy(x[None]).to(device)).cpu().numpy()[0, :, 0]
    pred_p = pred_n * sd + mu                                     # de-normalize

    h0, h1 = start + INPUT_LEN, start + INPUT_LEN + OUTPUT_LEN
    true_p = p[h0:h1].astype(float)

    # 2) MANUAL knock calculation, applied to the forecast and (for reference) the truth
    if sigma is None:
        sigma = baseline_sigma(pressure)
    pred_knock = knock_from_pressure(pred_p, start_index=h0, sigma=sigma)
    true_knock = knock_from_pressure(true_p, start_index=h0, sigma=sigma)

    idx = np.arange(h0, h1)
    fig, ax = plt.subplots(figsize=(14, 4))
    ax.plot(np.arange(start, h0), p[start:h0], color='lightgray', lw=0.8, label='input context')
    ax.plot(idx, true_p, color='steelblue', lw=1.3, label='true pressure')
    ax.plot(idx, pred_p, color='darkorange', lw=1.3, ls='--', label='predicted pressure')
    ax.scatter(idx[true_knock], true_p[true_knock], color='red', s=40, zorder=5,
               label='knock (from true signal)')
    ax.scatter(idx[pred_knock], pred_p[pred_knock], color='purple', marker='x', s=55, zorder=6,
               label='knock (from prediction)')
    ax.axvline(h0, color='gray', ls=':', lw=1)
    ax.set_xlabel('serialized sample index  (1 per crank angle)')
    ax.set_ylabel('knock-sensor pressure signal')
    ax.set_title(f'forecast @ start={start}   horizon MSE={np.mean((pred_p - true_p) ** 2):.3f}')
    ax.legend(loc='upper right', fontsize=8)
    plt.tight_layout()
    plt.show()
    return fig
```

```python
# Pick a validation window whose forecast horizon contains knock in the TRUE signal,
# so the overlay is easy to read. (Knock here is computed with the same manual rule.)
split = int(len(pressure) * (1 - VAL_FRAC))
sigma = baseline_sigma(pressure)
knock_all = knock_from_pressure(pressure, start_index=0, sigma=sigma)   # manual, full signal
kidx = np.where(knock_all)[0]
kidx = kidx[(kidx >= split + INPUT_LEN) & (kidx < len(pressure) - OUTPUT_LEN)]
start = int(kidx[len(kidx) // 2]) - INPUT_LEN if len(kidx) else split

plot_forecast(model, pressure, mu, sd, start=start, sigma=sigma)
```
